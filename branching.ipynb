Hier ein detaillierter Workflow, wie 20 Mitarbeitende im Team ohne Beeinträchtigung des Main-Branches gemeinsam an dem Projekt arbeiten können:

---

### 1. Repository klonen

Jeder Mitarbeiter klont das Repository lokal, z. B. so:

```bash
git clone https://github.com/maximsqrt/PYTHON_BASICS_BATCH202411_GITCHECK.git
```

Dann wechselt er in das Verzeichnis:

```bash
cd PYTHON_BASICS_BATCH202411_GITCHECK
```

---

### 2. Main-Branch aktuell halten

Bevor jemand an einer neuen Funktion arbeitet, sollte er den aktuellen Stand von `main` abrufen:

```bash
git checkout main
git pull origin main
```

---

### 3. Neuen Branch erstellen

Jeder Mitarbeiter erstellt einen eigenen Feature-Branch, um isoliert zu arbeiten. Dabei sollte ein aussagekräftiger Name gewählt werden:

```bash
git checkout -b feature/DEIN_FEATURE_NAME
```

Beispiel:

```bash
git checkout -b feature/neues-reporting
```

---

### 4. Änderungen vornehmen und committen

Im neuen Branch können nun Änderungen vorgenommen werden. Nachdem die Änderungen fertig sind, werden sie zum Staging-Bereich hinzugefügt und committet:

```bash
git add .
git commit -m "Beschreibung der vorgenommenen Änderung"
```

---

### 5. Branch zum Remote-Repository pushen

Der neue Branch wird zum Remote-Repository gepusht:

```bash
git push -u origin feature/DEIN_FEATURE_NAME
```

---

### 6. Pull Request (PR) erstellen

Nachdem der Branch online ist, erstellt der Mitarbeiter über die GitHub-Weboberfläche einen Pull Request von seinem Feature-Branch zum `main`-Branch. Dabei können Reviewer den Code überprüfen und Kommentare abgeben.

---

### 7. Merge in Main (nach Review)

Sobald der Pull Request von den Verantwortlichen freigegeben wurde, wird der Branch in den `main`-Branch gemerged. Dies sollte über die GitHub-Oberfläche erfolgen, um Konflikte und Versionsprobleme zentral zu lösen.

---

### Zusammenfassung des Workflows

- **Klonen:** Jeder Mitarbeiter klont das zentrale Repository.
- **Aktualisieren:** Vor jeder neuen Arbeit den `main`-Branch updaten.
- **Eigenen Branch erstellen:** Für jede neue Funktion oder Änderung einen eigenen Branch anlegen.
- **Commit & Push:** Änderungen im Branch committen und pushen.
- **Pull Request:** Änderungen über einen PR in den `main`-Branch mergen lassen.

Dieser Workflow stellt sicher, dass der `main`-Branch stabil bleibt und alle Teammitglieder unabhängig an ihren Features arbeiten können, ohne sich gegenseitig zu behindern.
